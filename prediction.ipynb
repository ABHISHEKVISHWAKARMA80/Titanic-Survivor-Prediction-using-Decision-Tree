```python
import numpy as np
import pandas as pd
from sklearn.preprocessing import LabelEncoder
from sklearn.tree import DecisionTreeClassifier
```

```python
data = pd.read_csv("Train.csv")
```

```python
data.head()
```

```python
data.info()
```

```python
columns_to_drop = ["name","ticket","cabin","embarked","boat","body","home.dest"]
```

```python
data_clean = data.drop(columns_to_drop,axis=1)
```

```python
data_clean.head()
```

```python
le = LabelEncoder()
data_clean["sex"] = le.fit_transform(data_clean["sex"])
```

```python
data_clean.head()
```

```python
#data_clean.info()
```

```python
data_clean = data_clean.fillna(data_clean["age"].mean())
```

```python
data_clean.info()
```

```python
input_cols = ['pclass',"sex","age","sibsp","parch","fare"]
output_cols = ["survived"]
X = data_clean[input_cols]
Y = data_clean[output_cols]
print(X.shape,Y.shape)
```

```python
sk_tree = DecisionTreeClassifier(criterion='entropy',max_depth=5)
```

```python
train_data = data_clean
test_data = pd.read_csv("Test.csv")
test_data.info()
```

```python
test_data = test_data.drop(columns_to_drop,axis=1)
```

```python
test_data["sex"] = le.fit_transform(test_data["sex"])
```

```python
test_data = test_data.fillna(test_data["age"].mean())
```

```python
test_data.info()
```

```python
train_data[input_cols]
```

```python
train_data[output_cols]
```

```python
test_data
```

```python
sk_tree.fit(train_data[input_cols],train_data[output_cols])
```

```python
sk_predict = sk_tree.predict(test_data)
```
